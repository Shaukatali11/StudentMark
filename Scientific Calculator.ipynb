```python
import tkinter as tk
import math

class CalculatorApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Scientific Calculator")

        self.result_var = tk.StringVar()
        self.result_var.set("")

        self.entry = tk.Entry(master, textvariable=self.result_var, font=('Arial', 16), bd=10, insertwidth=4, width=14,
                              justify='right')
        self.entry.grid(row=0, column=0, columnspan=4)

        buttons = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('C', 4, 1), ('=', 4, 2), ('+', 4, 3),
            ('sqrt', 5, 0), ('^', 5, 1), ('sin', 5, 2), ('cos', 5, 3),
            ('tan', 6, 0), ('log', 6, 1), ('(', 6, 2), (')', 6, 3),
            ('Del', 7, 0),  # Added 'Del' button
        ]

        for (text, row, col) in buttons:
            tk.Button(master, text=text, padx=20, pady=20, font=('Arial', 14), command=lambda t=text: self.on_button_click(t))\
                .grid(row=row, column=col)

    def on_button_click(self, button_text):
        if button_text == '=':
            try:
                result = eval(self.result_var.get())
                self.result_var.set(result)
            except (SyntaxError, NameError, TypeError):
                self.result_var.set("Error")
        elif button_text == 'C':
            self.result_var.set("")
        elif button_text == 'sqrt':
            try:
                result = math.sqrt(float(self.result_var.get()))
                self.result_var.set(result)
            except ValueError:
                self.result_var.set("Error")
        elif button_text == 'Del':  # Added delete functionality
            current_text = self.result_var.get()
            self.result_var.set(current_text[:-1])  # Remove the last character
        else:
            current_text = self.result_var.get()
            self.result_var.set(current_text + button_text)


def main():
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()


if __name__ == "__main__":
    main()
```
